```python
import fitz
from os import listdir
from os.path import isfile, isdir, join

pdf_path = r"C:\Users\User\Documents\pretty paper\normal"
files = listdir(pdf_path)
data = []

for f in files:
  # 產生檔案的絕對路徑
  fullpath = join(pdf_path, f)
  # 判斷 fullpath 是檔案還是資料夾
  if isfile(fullpath):
    doc = fitz.open(fullpath)
    onefileData = {'title':doc.metadata['title']}
    print("檔案：", f)
    #doc = fitz.open(fullpath)
    #print ("number of pages: %i" % doc.pageCount)
    #print(doc.metadata)

    page1 = doc.loadPage(0)
    page1text = page1.getText("text")
    #print(page1text)
    data.append(onefileData)
  elif isdir(fullpath):
    print("資料夾：", f)
print(data)
```
