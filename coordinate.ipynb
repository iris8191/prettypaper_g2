```python
import fitz

doc = fitz.open('Toward 6G Internet of Things and the Convergence With RoF System.pdf')

worddd = []
for page in doc:
    words=page.get_text_words()
    for w in words:
        #位置資訊：fitz.Rect(w[:4])
        #w[4]：文字資訊
        #print(w[4])
        print(w)
        worddd.append(w)
#fitz.Rect(w[:4]),
#print(worddd)
```

```python
wordsize=[] #出現的數值（不重複）
for i in worddd:
    a = i[3]-i[1]
    if a not in wordsize:
        wordsize.append(a)
    else:
        continue
print(wordsize)
```

```python
#整篇論文出現字的數值（不重複）&分別對應到的所有字
tidy = {}

for j in wordsize:
    all_wordsize = []
    #dic = {j:all_wordsize}
    dic = [j,all_wordsize]
    
    for i in worddd:
        
        number = i[3]-i[1]
        word = i[4]
        if number == j:
            all_wordsize.append(word)
    #print(dic)
    tidy[dic[0]] = dic[1]
    
print(tidy) 
```

```python
all_in = []
for i in worddd:
    a = i[3]-i[1]
    all_in.append(a)
print(all_in)
```

```python
count_finish = {} #每種大小各有幾個字
for i in all_in:
    how_much = all_in.count(i)
    count_finish[i] = how_much
print(count_finish)
```

```python
#找最大的字體，猜測為標題（怕有首字最大，所以做了一個判斷式）
biggest = max(wordsize)
#print(biggest)

wow = tidy[biggest]
wow_count = len(wow)

if wow_count == 1:
    wow_del1 = wordsize.remove(biggest)
    wow_del2 = tidy.pop(biggest)
    biggest2 = max(wordsize)
    #print(biggest2)
    new = tidy[biggest2]
    title = " ".join(new)
    print(title)
        

else:
    title = " ".join(new)
    print(title)

#print(wow)
```
